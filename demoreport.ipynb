# Postgame Report
BYU vs. Gonzaga Postgame Report

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
# Data Import
game_data = pd.read_csv("./post_game_stats.csv")
```

### BYU BOX SCORE

```python
byu_stats = game_data[game_data.Team == 'BYU']
byu_stats
```

### GONZAGA BOX SCORE

```python
gonzaga = game_data[game_data.Team == 'Gonzaga']
gonzaga
```

### Game Leaders

```python
leader_stats = ['MIN','PTS', 'REB', 'AST', 'BLK']
```

```python
fig = plt.figure(figsize=(10,15))
for idx, stat in enumerate(leader_stats):
    stat_leaders = game_data.sort_values(by = stat, ascending=False).head(3)
    fig.add_subplot(int('32{}'.format(idx+1)))
    plt.bar(x = list(stat_leaders['Player']), height = list(stat_leaders[stat]))
    plt.title('{} Leader'.format(stat))
plt.tight_layout()
plt.show()
```
